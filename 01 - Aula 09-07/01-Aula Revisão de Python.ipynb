```python
# 1 - Fa√ßa um Programa que pe√ßa dois n√∫meros ao usu√°rio e imprima o maior deles.

lista = []
for c in range(2):
    lista.append(int(input(f'Digite o n√∫mero { c + 1} ¬∞: ')))

print(lista)
lista.reverse()
print(f'O maior valor digitado foi : {lista[0]}')
```

```python
# 2 - Fa√ßa um Programa que pe√ßa ao usu√°rio um valor e mostre na tela se o valor √© positivo ou negativo.

numero = int(input(f'Digite o n√∫mero inteiro: '))
verficador = type(numero)
print(verficador)


if numero > 0:
    print(f'O {numero} √© positivo')
elif numero == 0:
    print(f'O {numero} √© neutro')
elif numero < 0:
    print(f'O {numero} √© negativo')
```

```python
# 3  -  Escreva  um  programa  que  solicite  ao  usu√°rio  uma  frase  e  conte  quantas vogais (a, e, i, o, u) existem nessa frase.

a = []
e = []
i = []
o = []
u = []
consoante = []

frase = input('Digite uma palavra: ')
frase = frase.upper()



for p in frase:
        if p == 'A':
            a.append(1)
        elif p == 'E':
            e.append(1)
        elif p == 'I':
            i.append(1)
        elif p == 'O':
            o.append(1)
        elif p == 'U':
            u.append(1)
        else:
            consoante.append(1)

        dicionario={'A':sum(a),'E':sum(e),'I':sum(i),'O':sum(o),'U':sum(u) ,'Consoante':sum(consoante) }
print(dicionario)
```

```python
# 1 - Voc√™ √© um operador de caixa em um supermercado e precisa implementar um programa para registrar as compras dos clientes.

# O programa deve solicitar o nome e o pre√ßo de cada produto e somar o total da compra no final. O operador pode inserir o nome e o pre√ßo de cada 
#produto repetidamente at√© que ele deseje encerrar a compra, digitando um valor especial (por exemplo, -1). 
# Utilize o la√ßo de repeti√ß√£o while para
# permitir a inser√ß√£o cont√≠nua dos pre√ßos dos produtos at√© que o operador
# decida encerrar a compra.

c = 1
lista = []
valor_total = 0
while True:
    produto_nome = input(f'Digite o nome do Produto {c}:')
    porduto_valor = float(input(f'Digite o valor do Produto {produto_nome}:'))
    c += 1
    dicionario = {'Prodto':produto_nome,'Valor':porduto_valor}
    lista.append(dicionario)
    print('-'*100)
    print(lista)
        
    for y in lista:
        valor_total+= y['Valor']
    print(f'At√© Agora voc√™ Gastou R$: {valor_total}')
    valor_total = 0
    print('-'*100)

    parar = input('Digite S para parar as Compras: ')

    if parar == 'S':
        valor_total = 0
        for y in lista:
            valor_total+= y['Valor']
        print('Sistema Finalizado !!!!')
        print(f'Comprou um total de {c - 1} produtos')
        print(f'Voc√™ gastou no total üò≤R$: {valor_total}')
        break





```

```python
# 2 - Voc√™ foi contratado como desenvolvedor para criar um programa que verifique a senha digitada pelos usu√°rios. O programa deve solicitar ao usu√°rio que digite uma 
#senha e verificar se ela atende aos crit√©rios estabelecidos. Os crit√©rios s√£o os seguintes:
# ‚óè A senha deve ter no m√≠nimo 8 caracteres e no m√°ximo 12 caracteres.


maisculo = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
minusculo = 'abcdefghijklmnopqrstuvwxyz'
numeros = '0123456789'

while True:
    controle_caract = False
    controle_maiuscula = False
    controle_minuscula = False
    controle_numeros = False

    senha = input('Digite a senha: ')
    n_caract =len(senha)
    
    if n_caract <= 12 and n_caract>=8:
        controle_caract = True

    for c in senha:
        if c in maisculo:
            controle_maiuscula = True
        if c in minusculo:
            controle_minuscula = True
        if c in numeros:
            controle_numeros = True
    
    if controle_caract is True and controle_maiuscula is True and controle_minuscula is True and controle_numeros is True:
        print(f'Sua senha {senha} foi aprovada')
        break
    else:
        print(f'Sua senha {senha} n√£o foi aprovada')
```
